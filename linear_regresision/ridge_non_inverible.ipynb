```python
import numpy as np
import plotly.graph_objects as go
```

```python
X = np.array([[1,2],[2,4],[3,6],[4,8]])
y = np.array([5,9,13,17])
```

```python
x_with_intercept = np.hstack([np.ones((X.shape[0],1)),X])
```

```python
lamda_ridge = 0.1
I = np.eye(x_with_intercept.shape[1])
beta_ridge = np.linalg.inv(x_with_intercept.T @ x_with_intercept + lamda_ridge*I) @ x_with_intercept.T @ y
```

```python
beta_ridge
```

```python
x1_range = np.linspace(0, 50, 20)
x2_range = 2 * x1_range
x1_mesh, x2_mesh = np.meshgrid(x1_range, x2_range)
y_mesh = beta_ridge[0] + beta_ridge[1]*x1_mesh + beta_ridge[2]*x2_mesh

fig = go.Figure(data=[
    go.Scatter3d(
        x=X[:, 0], y=X[:, 1], z=y,
        mode='markers',
        marker=dict(size=5, color='red'),
        name='Data Points'
    ),
    go.Surface(
        x=x1_mesh, y=x2_mesh, z=y_mesh,
        opacity=0.5,
        name='Ridge Regression Plane'
    )
])

fig.update_layout(
    scene=dict(
        xaxis=dict(title='x1', range=[0, 20]),
        yaxis=dict(title='x2', range=[0, 20]),
        zaxis=dict(title='y', range=[0, 20])
    ),
    title="3D Ridge Regression Fit"
)

fig.show()
```
