```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
n_samples = 30
X_input = np.linspace(0,10,n_samples)
y = 2*X_input + 1 + np.random.randn(n_samples)
'''np.random.randn(n_samples) generates random numbers from a 
standard normal distribution — that means the numbers are drawn from 
a Gaussian distribution with a mean of 0 and a standard deviation of 1'''
```

```python
X_input = np.append(X_input,[15,18,20])
y = np.append(y,[80,90,100])
```

```python
X_input
```

```python
X_input = X_input.reshape(-1,1)
```

```python

X_input.shape
```

```python
X = np.hstack([np.ones(X_input.shape),X_input])
```

```python
X.shape
```

```python
beta_OLS = np.linalg.inv(X.T @ X) @ X.T @ y
lambda_ridge = 1500
I = np.eye(X.shape[1])#np.eye create an identity matrix of given size
beta_Ridge = np.linalg.inv(X.T @ X + lambda_ridge*I) @ X.T @ y
# @ is used in the palce op np.dot as it is easier to use and less
# messier and np.linag is the mode of numpy conating various funtions
# like inverse,determinant,eigen vectors,eigen value
```

```python
I.shape
```

```python
beta_OLS
```

```python
plt.figure(figsize=(10,10)) # figsize take tuples as width and size in inches
plt.scatter(X_input,y,c = 'blue',label = 'Data Points')
x_range = np.linspace(X_input.min(),X_input.max(),100)
y_beta_ols = x_range*beta_OLS[1] + beta_OLS[0]
y_beta_ridge = x_range*beta_Ridge[1] + beta_Ridge[0]
plt.plot(x_range,y_beta_ols,c = 'red',label = 'Without Regularization')
plt.plot(x_range,y_beta_ridge,c = 'green',label = 'Regularization')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.legend()
plt.show()
```
