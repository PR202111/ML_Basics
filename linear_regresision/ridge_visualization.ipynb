```python
import numpy as np
import plotly.graph_objects as go
```

```python
np.random.seed(42)
```

```python
n_samples = 100
x1 = np.random.rand(n_samples)
x2 = 2*x1
```

```python
y = 3*x1 + 2*x2 + np.random.rand(n_samples)*0.5
```

```python
beta1_range = np.linspace(-10,10,200)
beta2_range = np.linspace(-20,20,200)
beta1,beta2 = np.meshgrid(beta1_range,beta2_range)
```

```python
def calculate_loss(beta1,beta2,x1,x2,y,lambda_ridge = 0):
    loss = np.zeros(beta1.shape)
    for i in range(beta1.shape[0]):
        for j in range(beta1.shape[1]):
            y_pred = beta1[i,j]*x1 + beta2[i,j]*x2
            loss[i,j] = np.sum((y - y_pred)**2) + lambda_ridge*(beta1[i,j]**2 + beta2[i,j]**2)
    return loss
```

```python
loss_ols = calculate_loss(beta1,beta2,x1,x2,y)
loss_ridge = calculate_loss(beta1,beta2,x1,x2,y,lambda_ridge=500)
```
