```python
def sigmoid(z):
    z = np.clip(z,-500,500)
    return 1/(1+np.exp(-z))
```

```python
def loss(theta,x,y,lam):
    m = len(y)
    z = np.dot(x,theta)
    hx = sigmoid(z)
    epsilon = 1e-9
    hx = np.clip(hx,epsilon,1-epsilon)
    l = -(y*np.log(hx) + (1-y)*np.log(1-hx))
    reg = lam*np.sum(theta[1:]**2)/2
    return (l+reg)/m
```

```python
def grad_decent(theta,x,y,lam,alpha,iters):
    m = len(y)
    cost = []
    theta_tilda = theta.copy()
    theta_tilda[0] = 0

    for i in range(iters):
        hx = sigmoid(np.dot(x,theta))
        error = hx - y

        grad = np.dot(x.T,error) + lam*theta_tilda/m

        theta -= alpha*grad
        cost.append(loss( theta,x, y, lam))

    
    return theta,cost
```
