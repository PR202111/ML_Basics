```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([[1,1],[1,-1],[-1,-1],[-1,1]])
y = np.array([1,-1,1,-1])
```

```python
plt.scatter(x[y==1][:,0],x[y==1][:,1],label='Positive',c='blue')
plt.scatter(x[y==-1][:,0],x[y==-1][:,1],label='Negative',c='red')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.legend()
plt.axhline(0,c='black',linewidth='1')
plt.axvline(0,c='black',linewidth='1')
plt.grid(True)
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

x_transformed = np.hstack((x, (x[:, 0] * x[:, 1]).reshape(-1, 1)))

# Create a figure
fig = go.Figure()

# Add points to the figure
colors = ['red' if label == 1 else 'blue' for label in y]
fig.add_trace(go.Scatter3d(x=x_transformed[:, 0], y=x_transformed[:, 1], z=x_transformed[:, 2],
                           mode='markers', marker=dict(color=colors, size=5)))

# Define separator lines
xx, yy = np.meshgrid(np.linspace(-2, 2, 50), np.linspace(-2, 2, 50))
zz = 0 * xx
fig.add_trace(go.Surface(x=xx, y=yy, z=zz, opacity=0.5, colorscale='greens'))

fig.update_layout(title='Transformed OR Points in 3D with separator',
                  scene=dict(xaxis_title='x1', yaxis_title='x2', zaxis_title='x3=x1*x2'),
                  margin=dict(l=0, r=0, b=0, t=30))

fig.show()
```
