```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1/(1+ np.exp(-z))
```

```python
x = np.array([-2,-1,1,2])
y = np.array([0,0,1,1])
```

```python
parameters = np.array([
    (1,0),   # theata = 1,theta0 = 0
    (-1,0),  # theata = -1,theta0 = 0
    (0.5,0), # theata = 0.5,theta0 = 0
    (5,0)    # theta = 5, theta0 = 0 ,likely the correct classifier
])
```

```python
# fig – Represents the overall figure.
# axs – Represents the array of subplot axes.
fig, axs = plt.subplots(2,2,figsize = (10,8))
# axs.ravel() – Flattens the 2D grid of
# subplots into a 1D array for easy iteration.
for ax, (theta,theta0) in zip(axs.ravel(),parameters):
    x_range = np.linspace(-3,3,300)
    z = theta*x_range + theta0
    y = sigmoid(z)

    # Plot the sigmoid function
    ax.plot(x_range,y,label = f"Theta = {theta},Theta0 = {theta0}")

    # Plot the points
    ax.scatter(x[:2],x[:2]-x[:2],c='red',label = 'Class -1')
    ax.scatter(x[2:],x[2:]-x[2:],c='blue',label = 'Class +1')

    # Formatting the plot
    ax.axhline(y=0.5,c='gray',linestyle = '--')
    ax.set_ylim(-0.1,1.1)
    ax.legend()

# Set overall figure title
fig.suptitle("Sigmoid Function for Differener Theta and Theta0 values")
plt.tight_layout(rect=[0,0.03,1,0.95])
plt.show()

    
```
