```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
# Building Dummy data
data={
    'Employee id': [10,20,15,25,30],
    'Gender': ['M','F','F','M','F'],
    'Remark': ['Good','Nice','Good','Great','Nice']
}
```

```python
# Converting the given data to data frame
df = pd.DataFrame(data)
```

```python
df
```

```python
#extract categorical columns froom the data frame
#here we extract the columns with object i.e (strings) datatype as they are the categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()

# Initialize OneHotdecoder
encoder = OneHotEncoder(sparse_output=False)

# Apply one hot encoding to the categorial columns
one_hot_encoded = encoder.fit_transform(df[categorical_columns])
# print(one_hot_encoded)
# print('\n')
#Create dataframe with the one hot encoded columns
# we use get_feature_names_out() to get the coulmns names for the encoded data
one_hot_df = pd.DataFrame(one_hot_encoded,columns=encoder.get_feature_names_out(categorical_columns))
# print(one_hot_df)

#Concatenate the one - hot encoded dataframe with the original dataframe
df_encoded = pd.concat([df,one_hot_df],axis=1)
# print('\n')
# print(df_encoded)
#Drop the original categorical coulmns
df_encoded = df_encoded.drop(categorical_columns,axis=1)
# print('\n')
print(df_encoded)
#Display the resulting dataframe
#print(df_encoded)
#print(categorical_columns)




```
