```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# defining data points
x = np.array([1,2,-1,-2])
y = np.array([1,1,0,0])
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
theta = np.array([1,-1,2,-2])
theta0 = np.array([0,1,-1,2])
```

```python
def cross_entropy_loss(x,y,theta,theta0):
    loss = 0
    for x_val,y_val in zip(x,y):
        z = sigmoid(np.dot(theta,x_val)+theta0)
        loss += -((y_val)*np.log(z)+(1-y_val)*np.log(1-z))
    return loss
```

```python
fig,axs = plt.subplots(2,2,figsize=(14,10))
fig.suptitle('Sigmoid function and cross entropy loss')

for ax,theta_v,theta0_v in zip(axs.flatten(),theta,theta0):
    x_range = np.linspace(-3,3,300)
    z = theta_v*x_range + theta0_v
    probability = sigmoid(z)

    ax.plot(x_range,probability,label=f'Sigmoid theta ={theta_v}, theta0 = {theta0_v}')

    ax.scatter(x,np.zeros_like(x),c = ['blue' if l == 1 else 'red' for l in y],s = 50)
    ax.vlines(x,0,sigmoid(theta_v*x + theta0_v),color = 'green' , linestyle = 'dotted')
    ax.hlines(y=0.5,xmin=-3,xmax=3,color='grey',linestyles='dotted')
    loss = cross_entropy_loss(x,y,theta_v,theta0_v)
    ax.title.set_text(f'Theta = {theta_v},Theta0 = {theta0_v}, Loss = {loss:.2f}')

    ax.set_xlim([-3,3])
    ax.set_ylim([-0.1,1.1])
    ax.legend()

plt.tight_layout(rect=[0,0.03,1,0.95])
plt.show()



```
