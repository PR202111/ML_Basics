```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2
```

```python
def df(x):
    return 2*x
```

```python
x_start = 9
learning_rate = 0.15
n_iterations = 10
```

```python
x_steps = [x_start]
y_steps = [f(x_start)]
```

```python
x = np.linspace(-10,10,400)
y = f(x)
```

```python
for i in range (n_iterations):
    plt.figure(figsize=(8,5))

    current_x = x_steps[-1]
    current_y = y_steps[-1]

    gradient = df(current_x)

    next_x = current_x - learning_rate*gradient
    x_steps.append(next_x)
    y_steps.append(f(next_x))

    plt.plot(x,y,c = 'black',label = 'f(x) = x^2')
    plt.scatter(x_steps,y_steps,c = 'green')
    plt.plot(x_steps,y_steps,linestyle='dashed',marker = 'o',c='blue',label = 'Gradient decent path')

    tangent_x = np.linspace(current_x-1,current_x+1,10)
    tangent_y = gradient*(tangent_x-current_x) + current_y
    plt.plot(tangent_x,tangent_y,linestyle = 'dashed',c = 'red',linewidth = 0.9,label = 'Tangent at current step')

    plt.title(f'Gradient Decent on f(x) = x^2 (Iteration {i+1})')
    plt.xlim(-10,10)
    plt.ylim(-10,100)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid()
    plt.show()
```
